```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(rc={'figure.figsize':(7.7,7.7)}) # graph size
```

```python
with open("student_aggregate.json", "r") as f:
    data_s_aggregate = json.load(f)
    
df_aggregate = pd.DataFrame(data_s_aggregate) 
df_aggregate.replace('', np.nan, inplace=True)

print('shape - df_aggregate', df_aggregate.shape)
print(df_aggregate.info())
df_aggregate.head()
```

```python
for column in df_aggregate:
    print(column)
    print(df_aggregate[column].value_counts())
    print('**********************************')
```

```python
df_aggregate.shape
```

```python
id_unicos = df_aggregate['Id'].drop_duplicates().dropna()
len(id_unicos)
```

# tempo de cadastro

```python
for i in {'LastLoginDate', 'RegisteredDate'}:
    df_aggregate[i] = pd.to_datetime(df_aggregate[i])
```

```python
df_aggregate['tempo_user'] = df_aggregate['LastLoginDate'] - df_aggregate['RegisteredDate']
```

```python
df_aggregate['tempo_user'] = df_aggregate['tempo_user'].dt.days
df_aggregate['tempo_user'] = df_aggregate['tempo_user'].astype(float)
df_aggregate.tempo_user = df_aggregate.tempo_user.replace(-1, 0)
```

```python
df_aggregate[['LastLoginDate', 'RegisteredDate', 'tempo_user']].head()
```

```python
df_aggregate['tempo_user'].value_counts()
```

# estados com mais alunos

```python
df_states = df_aggregate['stateName'].value_counts(normalize=True).reset_index()
```

```python
df_states.columns = ['estado', 'porcentagem']
df_states.head()
```

```python
df_states['porcentagem'] = df_states['porcentagem']*100 
```

```python
df_states.head()
```

```python
graph_estados = sns.barplot(x='estado', y='porcentagem', data=df_states, palette='YlOrBr_r')
for tick in graph_estados.get_xticklabels():
    tick.set_rotation(90)

plt.xlabel('')
plt.ylabel('% de acesso')
plt.title('')
plt.tight_layout()
graph_estados.figure.savefig('porc_estado.png', dpi=100)
```
