```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(rc={'figure.figsize':(7.7,7.7)}) # graph size
```

```python
with open("student_answer.json", "r") as f:
    data_s_answer = json.load(f)
    
df_answer = pd.DataFrame(data_s_answer['student_answer']) 
df_answer.replace('', np.nan, inplace=True)

print('shape - df_answer', df_answer.shape)
print(df_answer.info())
df_answer.head()
```

```python
df_answer.head()
```

```python
for i in {'Active', 'AnswerClient', 'AnswerId', 'AnswerQuestionId', 'Client', 'Gender',
          'IsMobileRegister', 'UploadCount'}:
    print(i)
    print(df_answer[i].value_counts())
    print('******************************************************')
```

# Em qual device o consumo é maior?

```python
df_answer.Client=np.where(df_answer.Client.isin(['Website']),df_answer.Client,'Mobile')
```

```python
tipo_dispositivo = df_answer['Client'].value_counts(normalize=True).reset_index()
tipo_dispositivo.columns = ['local_acesso', 'porcentagem']
```

```python
g_Client = sns.barplot(x='local_acesso', y='porcentagem', data=tipo_dispositivo, palette='YlOrBr')
g_Client.set(xlabel='', ylabel='% dos acessos');

g_Client.figure.savefig('porcentagem_acessos.png', dpi=100)
```

# Perguntas respondidas por Id

```python
df_perguntas_id = df_answer[['Id', 'AnswerQuestionId']].drop_duplicates().dropna()
df_perguntas_id.shape
```

```python
df_perguntas_id = df_perguntas_id.groupby('Id')['AnswerQuestionId'].count().reset_index()
df_perguntas_id.sort_values(by='AnswerQuestionId', ascending=False, inplace=True)
df_perguntas_id.head()
```

```python
# investigacao 
df_answer[df_answer.Id == '6683737']
```

```python
print('media', df_perguntas_id['AnswerQuestionId'].mean())
print('mediana', df_perguntas_id['AnswerQuestionId'].median())
```

```python
plt.hist(df_perguntas_id['AnswerQuestionId'], alpha=0.7, color='orange', bins=75)

plt.xlabel('perguntas respondidas por usuario')  
plt.ylabel('frequencia')
plt.xlim(0,50)
plt.savefig('hist_perguntas_user.png', dpi=100);
```

# Tempo usando o Passei Direto

```python
df_answer.head()
```

```python
df_answer.dtypes
```

```python
for i in {'LastLoginDate', 'RegisteredDate'}:
    df_answer[i] = pd.to_datetime(df_answer[i])
```

```python
df_answer['tempo_user'] = df_answer['LastLoginDate'] - df_answer['RegisteredDate']
```

```python
df_answer[['LastLoginDate', 'RegisteredDate', 'tempo_user']].head()
```

```python
df_answer.dtypes
```

```python
df_answer['tempo_user'] = df_answer['tempo_user'].dt.days
```

```python
unique_ids = df_answer['Id'].drop_duplicates().dropna()
len(unique_ids)
```

```python
df = df_answer.groupby('Id')['tempo_user'].max().reset_index()
df.head()
```

```python
df.shape
```

```python
df.sort_values(by='tempo_user', ascending=False, inplace=True)
df.head()
```

```python
print('media', df_answer['tempo_user'].mean())
print('mediana', df_answer['tempo_user'].median())
```

```python
plt.hist(df_answer['tempo_user'], alpha=0.7, color='orange')

plt.xlabel('tempo de cadastro do usuario (em dias)')  
plt.ylabel('frequencia')
plt.savefig('hist_tempo_user.png', dpi=100);
```
